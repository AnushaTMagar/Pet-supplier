```python
import pandas as pd
```

```python
data = pd.read_csv('C:\\Users\\anush\\OneDrive\\Desktop\\pet_supplies_2212.csv')
```

```python
column_names = data.columns

print(column_names)
```

```python

data['category'] = data['category'].replace('-', 'Unknown')
count_missing = data['category'].str.count('Unknown').sum()

print(f"Total occurrences of missing values: {count_missing}")
print(data.head(25))
```

```python
#convert all datasize to lowercase.
data['size'] = data['size'].str.lower()
print(data.head(25))
```

```python
count_missing = data['price'].str.count('unlisted').sum()
print(f"Total occurrences of missing values: {count_missing}")

# Convert non-numeric values to NaN in the "price" column
data['price'] = pd.to_numeric(data['price'], errors='coerce')

# Calculate the overall median price
overall_median_price = data['price'].median()
print(f"Overall median price: {overall_median_price}")

```

```python
# Replace missing values in the "price" column with the overall median price
data['price'].fillna(overall_median_price, inplace=True)

# Round values in the "price" column to two decimal places
data['price'] = data['price'].round(2)

# Print the first few rows to verify the changes
print(data.head(25))
```

```python
# Convert the "sales" column to numeric, coercing non-numeric values to NaN
data['sales'] = pd.to_numeric(data['sales'], errors='coerce')

# Count non-numeric values
non_numeric_count = data['sales'].isnull().sum()

print(f"Non-numeric values count in 'sales' column: {non_numeric_count}")
```

```python
overall_median_sales = data['sales'].median()
print(f"Overall median sales: {overall_median_sales}")
```

```python
# Convert the "sales" column to numeric, coercing non-numeric values to NaN
data['rating'] = pd.to_numeric(data['rating'], errors='coerce')

# Count non-numeric values
non_numeric_count = data['rating'].isnull().sum()

print(f"Missing values count in 'rating' column: {non_numeric_count}")
```

```python
data['rating'].fillna(0, inplace=True)
data['rating'] = data['rating'].astype(int)
print(data.head(25))
```

```python
# Count missing values in the "repeat_purchase" column
missing_values_count = data['repeat_purchase'].isnull().sum()
print(f"Count of missing values in 'repeat_purchase' column: {missing_values_count}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#a bar plot to show the count of repeat purchases by category
plt.figure(figsize=(10, 6))
ax = sns.countplot(data=data, x='category', hue='repeat_purchase')

# Add count labels on top of the bars
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height), ha='center', va='bottom')

plt.title("Count of Repeat Purchases by Category")
plt.xlabel("Category")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title='Repeat Purchase', labels=['Not Repeat', 'Repeat'])
plt.tight_layout()

# Show the plot
plt.show()
plt.show()
```

```python
# Create a stacked bar plot to show the proportions of repeat purchases by category
plt.figure(figsize=(10, 6))
ax = sns.countplot(data=data, x='category', hue='repeat_purchase')

# Add count labels on top of the bars
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height), ha='center', va='bottom')

plt.title("Proportions of Repeat Purchases by Category")
plt.xlabel("Category")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title='Repeat Purchase', labels=['Not Repeat', 'Repeat'])
plt.tight_layout()

# Show the plot
plt.show()
```
